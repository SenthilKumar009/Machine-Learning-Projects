### Filtering - Using Conditional to Filter Cows and Columns

```python
df[df['Country'] == 'India']

# df['Country'] == 'India' -> This will return the true/ false values which indicates, 
# how many columns india (True) and not contains india (False)

# df[df['Country'] == 'India'] -> Will return the records which contains only True value
```

```python
# Store the filter condition in a variable and apply on the dataset
filt = (df['Country'] == 'India')
df[filt]
```

```python
# Using loc method
df.loc[filt].head()
```

```python
df.loc[filt, ['MainBranch', 'CurrencySymbol']]
```

**Filter using & and | condition**

```python
filt = (df['Country'] == 'India') & (df['CurrencySymbol'] == 'INR')
df[filt].head()
```

```python
# To display the records other than the filtered values
filt = (df['Country'] == 'India') & (df['CurrencySymbol'] == 'INR')
df[~filt]
```

```python
# Retuen the employees who are all getting the max salary
#df['ConvertedComp'].max()
filt = df['ConvertedComp'] == df['ConvertedComp'].max()
df[filt]
```
